# 🧪 Exploration du Churn Client

Ce notebook analyse les données de Telco pour identifier pourquoi les clients quittent l'entreprise.

### Importation des bibliothèques et chargement des données
Nous importons les outils nécessaires (Pandas pour les données, Seaborn/Matplotlib pour les graphiques) et chargeons notre fichier CSV.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des données
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

### 1. Quel est le taux de churn global ?
Nous calculons la proportion de clients ayant quitté l'entreprise pour avoir une vue d'ensemble de la situation.

```python
plt.figure(figsize=(7,7))
df['Churn'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Répartition Globale du Churn (Impact Business)')
plt.ylabel('')
plt.show()
```

### 2. Quel type de contrat churne le plus ?
Nous comparons les types de contrats pour voir si l'engagement longue durée réduit le risque de départ.

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Contract', hue='Churn', palette='viridis')
plt.title('Impact du Type de Contrat sur la Fidélité Client')
plt.show()
```

### 3. Les clients qui partent paient-ils plus cher ?
Nous analysons les charges mensuelles (MonthlyCharges) pour voir si le prix est un facteur de départ.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='magma')
plt.title('Comparaison des Frais Mensuels : Clients Fidèles vs Partis')
plt.show()
```

### 4. Après combien de mois en moyenne un client part-il ?
Nous calculons la moyenne de l'ancienneté (tenure) uniquement pour les clients qui ont churné.

```python
avg_tenure_churn = df[df['Churn'] == 'Yes']['tenure'].mean()
print(f"Ancienneté moyenne au moment du départ : {avg_tenure_churn:.1f} mois")

plt.figure(figsize=(10,6))
sns.histplot(data=df[df['Churn'] == 'Yes'], x='tenure', kde=True, color='red')
plt.title('Moment Critique du Départ (Ancienneté en mois)')
plt.show()
```

### 5. Quelles variables sont les plus liées au churn ?
Nous transformons le Churn en valeur numérique pour calculer les corrélations avec les autres variables numériques.

```python
df_numeric = df.copy()
df_numeric['Churn_num'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
correlations = df_numeric.corr(numeric_only=True)['Churn_num'].sort_values(ascending=False)

plt.figure(figsize=(10,6))
correlations.drop('Churn_num').plot(kind='bar', color='skyblue')
plt.title('Facteurs Numériques les plus corrélés au Churn')
plt.show()
```

## 📈 Conclusions Business

1. **Le prix est un levier majeur** : Les clients qui partent paient en moyenne des mensualités plus élevées que ceux qui restent.
2. **Le contrat mensuel est à haut risque** : La grande majorité des départs provient des contrats sans engagement ("Month-to-month").
3. **Le risque est immédiat** : Un grand nombre de clients partent dès les premiers mois. Il est crucial de renforcer l'onboarding pour stabiliser les nouveaux clients.
